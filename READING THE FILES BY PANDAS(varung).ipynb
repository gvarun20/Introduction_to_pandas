```python
import xml.etree.ElementTree as ET
import pandas as pd

# Load the XML file
tree = ET.parse('pandas.xml')
root = tree.getroot()

# Define a dictionary to store data
data = {"Name": [], "Age": [], "City": []}

# Iterate through the XML tree and extract data
for person in root.findall('.//tr'):
    name = person.find('td[@class="name"]').text
    age = person.find('td[@class="age"]').text
    city = person.find('td[@class="city"]').text

    data["Name"].append(name)
    data["Age"].append(age)
    data["City"].append(city)

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
